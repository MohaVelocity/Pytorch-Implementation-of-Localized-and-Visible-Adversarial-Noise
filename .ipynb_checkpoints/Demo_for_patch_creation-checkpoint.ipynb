<font size="6">Demo for creating a transferable adversarial patch </font>  
The resulting patch is location transferable and image transferable.  
The model used for creating the adversarial patch is a pretrained Inception-V3.  
10 images are randomly selected from the Imagenet-1K for training.

```python
import numpy as np
import os
import glob
from PIL import Image
import PIL 
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F
from torch.autograd import Variable
from torchvision import datasets, transforms, models

from utils import *
```

```python
#Checking if a GPU with cuda capability is available
device = (torch.device('cuda') if torch.cuda.is_available()
          else torch.device('cpu'))
print(f"Training on device {device}.")
```

<font size="4">Loading the pretrained model and the training Dataset </font> 

```python
model,preprocess = load_model('inception_v3',device)
path = os.path.join(os.getcwd(),'images')
cropped_images,preprocessed_images,label_list = load_and_preprocess_images(path,device,preprocess)
```

```python
#Displaying all images used for training
for i in range(len(preprocessed_images)):
    plt.subplot((len(preprocessed_images)//5) + 1,5,i+1)
    plt.axis('off')
    plt.title("Image: {}".format(i+1))
    plt.imshow(cropped_images[i])
plt.tight_layout()
plt.show()
```

```python
#Displaying with what probability the model guesses the correct label
print_nonadversarial_result(preprocessed_images,model,label_list)
```

<font size="4">Defining the patch size and training the patch </font> 

```python
#Setting the size of the adversarial patch
#The patch is in a square shape of 50x50 in this example
patch_size = 60
#Defining how many times the image is updated through back propagation using all training images
epochs = 10
#How many gradient descent steps to make for each image in an epoch
iter_num = 10
#How large each step is when trained with a specific image at a random location
epsilon = 0.1
#The minimum and maximum value of pixel values in the afverserial patch
clamp_val = 2
#The desired output that the patch is tricking the model to output
target = 855

#Generate the patch
patch = generate_patch(model,preprocessed_images,label_list,target,epsilon,epochs,iter_num,patch_size,device,clamp_val)
```

<font size="4">Showing and saving the results </font>

```python
#Displaying the probability for the original correct label and the target label after applying the patch
print_numeric_results(model,preprocessed_images,label_list,patch,target,device)    
```

```python
#Unormalize the patch to range [0,1] by undoing the preprocessing
unormalize = UnNormalize(mean=preprocess.mean,std=preprocess.std)
unorm_patch = unormalize(patch[0])
```

```python
#Transform the patch to PIL image and numpy array and display the PIL image
transform = transforms.ToPILImage()
patch_img = transform(unorm_patch)
patch_np = unorm_patch.cpu().detach().numpy()
plt.axis('off')
plt.imshow(patch_img)

```

```python
#Save the results as a .npy file and as .png file
path = os.path.join(os.getcwd(),"output_patch",str(target))
patch_img.save(path+".png") 
np.save(path+".npy",patch_np)
```
